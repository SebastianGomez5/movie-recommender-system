```python
credits = credits.rename(columns={'movie_id': 'id'})
```

```python
# unir ambas tablas
movies = movies.merge(credits, on='id')
movies
```

```python
#Eliminar informacion con valores nulos
movies.isnull().sum()
```

```python
movies = movies.dropna()

movies.isnull().sum()
movies
```

```python
#elegir columnas importantes
movies = movies[['id', 'original_title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
movies.head(1)
```

```python
import ast # Para convertir cadenas de texto en listas
# Función para convertir texto en lista
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies['cast'] = movies['cast'].apply(convert)
movies['crew'] = movies['crew'].apply(convert)
```

```python
movies
```

```python
#Solo mostrar los primeros 3 actores del cast
movies['cast'] = movies['cast'].apply(lambda x: x[:3])
movies.head(2)
```

```python
movies.sample(5)
```

```python
#Aplicar la operacion a la columna overview para dividir cada resumen en palabras individuales
movies['overview'] = movies['overview'].apply(lambda x: x.split())
movies.sample(5)
```

```python
# funcion para eliminar espacios 
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ", ""))
    return L1


movies['keywords'] = movies['keywords'].apply(collapse)
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)

movies.head(1)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies
```

```python
movies_new = movies[['id', 'original_title', 'tags']]
movies_new = movies_new.reset_index(drop=True) #Resetear indices
movies_new
```

```python
# hacer que tags no sea una lista
movies_new['tags'] = movies_new['tags'].apply(lambda x: ' '.join(x))
movies_new.head(3)
```

### Creacion del modelo de Maching Learning

```python
from sklearn.feature_extraction.text import CountVectorizer

#cv = CountVectorizer()
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = cv.fit_transform(movies_new['tags']).toarray() 
vector
```

Se busca transformar las palabras en numeros para que los modelos entienden numeros, vectorizandolas

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vector)
similarity
```

```python
movies_new[movies_new['original_title'] == 'The Lego Movie'].index[0]
```

```python
def recommend_movieV1(movie_title):
    # Verifica si la película existe
    if movie_title not in movies_new['original_title'].values:
        return f"La película '{movie_title}' no se encuentra en el dataset"
    movie_index = movies_new[movies_new['original_title'] == movie_title].index[0]
    similar_movies = get_similar_movies(movie_index)
    return similar_movies
def get_similar_movies(movie_index):
    # Get the similarity scores for the given movie
    sim_scores = similarity[movie_index]
    # Get the indices of the most similar movies
    similar_indices = sim_scores.argsort()[::-1][1:6]
    # Return the titles of the most similar movies
    return movies_new['original_title'].iloc[similar_indices].tolist()
```

```python
def recommend_movieV2(movie_title):
    movie_index = movies_new[movies_new['original_title'] == movie_title].index[0]
    distance = sorted(list((enumerate(similarity[movie_index]))), reverse=True, key=lambda x: x[1])
    return [movies_new['original_title'].iloc[i[0]] for i in distance[1:10]]
```
